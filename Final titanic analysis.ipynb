```python
df=pd.read_csv('titanic.csv')
df
```

# The key columns in the Titanic dataset:

###### PassengerId: A unique identifier for each passenger.
###### Survived: Indicates whether a passenger survived (1) or did not survive (0).
###### Pclass (Passenger Class): The class of the ticket the passenger purchased (1st, 2nd, or 3rd).
###### Name: The name of the passenger.
###### Sex: The gender of the passenger.
###### Age: The age of the passenger.
###### SibSp (Number of Siblings/Spouses Aboard): The count of siblings or spouses traveling with each passenger.
###### Parch (Number of Parents/Children Aboard): The count of parents or children traveling with each passenger.
###### Ticket: The ticket number.
###### Fare: The amount of money paid for the ticket.
###### Embarked: The port at which the passenger boarded the Titanic (C = Cherbourg, Q = Queenstown, S = Southampton).

# Data Visualization:

```python
#  create a histogram to visualize the distribution of passenger ages

plt.figure(figsize=(8, 6))
plt.hist(df['Age'].dropna(), bins=20, color='y', edgecolor='black')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# Filtering Data:

```python
# Filter the dataset to show only female passengers who survived

female_survivors = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
female_survivors
```

# Grouping and Aggregation:

```python
# Group by 'Pclass' and calculate the average fare
average_fare_by_class = df.groupby('Pclass')['Fare'].mean()
average_fare_by_class
```

# Data Cleaning:

```python
# Handle missing values in the 'Age' column.so 
# Fill missing values in 'Age' with the mean age

df['Age'].fillna(df['Age'].mean(),inplace=True)
df.head(10)
```

# Data Visualization:

```python
# Create a count plot to visualize the number of passengers in each passenger class

# Set the style for seaborn
sns.set(style="whitegrid")

# Create a count plot for passenger class
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', data=df, palette='Set2')
plt.title('Number of Passengers in Each Class')
plt.show()
```

#  Correlation Analysis:

```python
# Visualize a correlation between age and fare

# Calculate the correlation between 'Age' and 'Fare'
correlation_age_fare = df['Age'].corr(df['Fare'])

# Visualize the correlation with a scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.title(f'Correlation between Age and Fare: {correlation_age_fare:.2f}')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

# Age Distribution by Class:

```python
# Create a box plot to visualize the age distribution for each passenger class

# Create a box plot for age distribution by class
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Age', data=df, palette='pastel')
plt.title('Age Distribution by Passenger Class')
plt.show()
```

#  Survival Analysis:

```python
#  percentage of passengers who survived overall

# Calculate the overall survival rate
survival_rate =df['Survived'].mean() * 100
print(f"Overall Survival Rate: {survival_rate:.2f}%")
```

# Class Distribution by Gender:

```python
# Create a stacked bar chart for class distribution by gender
class_gender_distribution = df.groupby(['Pclass', 'Sex']).size().unstack()
class_gender_distribution.plot(kind='bar', stacked=True, colormap='viridis', figsize=(8,6))
plt.title('Class Distribution by Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

#  Fare Distribution by Passenger Class and Gender:

```python
# Create a violin plot to visualize the fare distribution based on passenger class and gender

plt.figure(figsize=(8,6))
sns.violinplot(x='Pclass', y='Fare', hue='Sex', data=df, split=True, palette='Set1')
plt.title('Fare Distribution by Passenger Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

#  Survival Rate by Class and Gender:

```python
# Calculate survival rate by class and gender
survival_rate_by_class_gender =df.groupby(['Pclass', 'Sex'])['Survived'].mean()

# Display the survival rate
print("Survival Rate by Class and Gender:")
survival_rate_by_class_gender
```

# Survival Rate by Age Group:

```python
# Analyze the survival rate based on different age groups

# Create age groups and calculate survival rate
age_bins = [0, 18, 30, 50, 100]
age_labels = ['0-18', '19-30', '31-50', '51+']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
survival_rate_by_age_group =df.groupby('Age_Group')['Survived'].mean()

# Display the survival rate by age group
print("Survival Rate by Age Group:")
survival_rate_by_age_group
```

# Fare Distribution by Embarked Port:

```python
# Analyse the fare distribution vary for passengers who embarked from different ports

# Create a violin plot for fare distribution by embarked port
plt.figure(figsize=(8,5))
sns.violinplot(x='Embarked', y='Fare', data=df, palette='coolwarm')
plt.title('Fare Distribution by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Fare')
plt.show()
```

# Fare Correlation Heatmap:

```python
# Create a correlation matrix
correlation_matrix =df.corr()

# Create a heatmap for the correlation matrix
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix,cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

# Embarked Port Analysis:

```python
# Distribution of passengers across different embarked ports

# Create a count plot for passengers by embarked port
plt.figure(figsize=(6,4))
sns.countplot(x='Embarked', data=df, palette='viridis')
plt.title('Distribution of Passengers by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Count')
plt.show()
```

# Age and Gender Analysis:

```python
# Create a swarm plot to visualize the age distribution by gender

# Create a swarm plot for age distribution by gender
plt.figure(figsize=(8,6))
sns.swarmplot(x='Sex', y='Age', data=df, palette='muted')
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

# Age and Fare Distribution Comparison:

```python
# Create a side-by-side comparison of the age and fare distributions using box plots

plt.figure(figsize=(8,4))
sns.boxplot(data=df[['Age', 'Fare']], palette='Set2')
plt.title('Comparison of Age and Fare Distributions')
plt.ylabel('Value')
plt.show()
```

# Passenger Class and Embarked Port Analysis:

```python
# Create a bar chart for passenger class distribution by embarked port
class_distribution_by_port =df.groupby(['Embarked', 'Pclass']).size().unstack()
class_distribution_by_port.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(8,6))
plt.title('Passenger Class Distribution by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Count')
plt.show()
```

# Family Size Analysis:

```python
# Distribution of family sizes (sum of 'SibSp' and 'Parch') among passengers

# Create a new column for family size
df['Family_Size'] =df['SibSp'] + df['Parch']

# Visualizing the distribution of family sizes
plt.figure(figsize=(8,6))
sns.countplot(x='Family_Size', data=df, palette='Set2')
plt.title('Distribution of Family Sizes')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.show()
```
